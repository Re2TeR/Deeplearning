## 주요 코드 미리 보기
----

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

X=df.iloc[:,3:]
sc=StandardScaler()
X=sc.fit_transform(X)
# print(X)
model = KMeans(n_clusters=2, random_state=42) # 지도 학습이 아니기 때문에 x에 대하여만 넣어주며 됨
y_preds=model.fit_predict(X)
df['cluster']=y_preds
print(f'실루엣 점수: {silhouette_score(X,y_preds):0.4f} ')
print()
df.head()
```

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='cluster')
```

## 코드 구현하기
----
   

```python
silhouette_avg=[]
print(X.shape)
for k in range(10):
  model=KMeans(n_clusters=k+2, random_state=42)
  y_preds=model.fit_predict(X)
  score=silhouette_score(X,y_preds)
  silhouette_avg.append(score)
  print(f'군집개수:{k+2}개, 평균 실루엣 점수: {score:.4f}')
```

```python
plt.plot(range(2,12),silhouette_avg, 'bo--')
plt.xlabel('# of clusters')
plt.ylabel('silhoutte_avg')
plt.show()
# k가 5정도 일대 실루엣 점수가 높음을 알 수 있음
```

```python
model=KMeans(n_clusters=5, random_state=42)
print(X.shape)
y_preds=model.fit_predict(X)
df['cluster']=y_preds
```

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='cluster', palette='deep')
```

```python
sns.barplot(x='cluster', y='Spending Score (1-100)', data=df)
```
